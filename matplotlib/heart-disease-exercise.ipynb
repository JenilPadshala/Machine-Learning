# Heart disease exercise

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
#Load dataset
heart_disease = pd.read_csv("../data/heart-disease.csv")
heart_disease.head()
```

```python
#create histogram for age column
heart_disease["age"].plot.hist();
```

```python
heart_disease.head()
```

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]> 50]
over_50.head()
```

```python
len(over_50)
```

```python
#pyplot method
over_50.plot(kind="scatter", x='age', y="chol", c='target')
```

```python
#OO method (not exactly ;)

fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);
# ax.set_xlim([50,80])
# ax.set_ylim([100,600])
```

```python
##OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
#customize the plot
ax.set(title="Age v. Cholestrol",
       xlabel = "Age",
       ylabel="Cholestrol")

#add legend
ax.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax.axhline(over_50['chol'].mean(), linestyle='--', color='red');
```

```python
over_50.head()
```

```python
#Subplot of chol, age, thalach

fig, (ax0,ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True)

#Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'])

#customize ax0
ax0.set(title='Age v Cholestrol', xlabel='age', ylabel='cholestrol')
ax0.legend(*scatter.legend_elements(), title="target")
ax0.axhline(y=over_50['chol'].mean(), linestyle='--', color='red')

#Add data to ax1
scatter1 = ax1.scatter(x=over_50['age'],
                       y=over_50['thalach'],
                       c=over_50['target'])

#customize ax1
ax1.set(title='Age v Thalach', xlabel='age', ylabel='thalach')
ax1.legend(*scatter.legend_elements(), title="target")
ax1.axhline(y=over_50['thalach'].mean(), linestyle='--', color='red')

#title the fig
fig.suptitle("Heart Disease Analysis", fontsize=16,fontweight='bold' )
```

```python
##OO method from scratch
plt.style.use('seaborn-whitegrid')
fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap='winter') # This changes colour scheme
#customize the plot
ax.set(title="Age v. Cholestrol",
       xlabel = "Age",
       ylabel="Cholestrol")

#add legend
ax.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax.axhline(over_50['chol'].mean(), linestyle='--', color='red');
```

```python
#Subplot of chol, age, thalach

fig, (ax0,ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True)

#Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap='winter')

#customize ax0
ax0.set(title='Age v Cholestrol', xlabel='age', ylabel='cholestrol')
ax0.legend(*scatter.legend_elements(), title="target")
ax0.axhline(y=over_50['chol'].mean(), linestyle='--', color='red')
ax0.set_xlim([50,80])
# ax0.set_ylim([])

#Add data to ax1
scatter1 = ax1.scatter(x=over_50['age'],
                       y=over_50['thalach'],
                       c=over_50['target'],
                       cmap='winter')

#customize ax1
ax1.set(title='Age v Thalach', xlabel='age', ylabel='thalach')
ax1.legend(*scatter.legend_elements(), title="target")
ax1.axhline(y=over_50['thalach'].mean(), linestyle='--', color='red')
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

#title the fig
fig.suptitle("Heart Disease Analysis", fontsize=16,fontweight='bold' )
```
